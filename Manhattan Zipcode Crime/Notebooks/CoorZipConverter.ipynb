```python
import pandas as pd
import shapefile
from shapely.geometry import Point
from shapely.geometry import shape
from tqdm import tqdm_notebook as tqdm
```

```python
shp = shapefile.Reader('ZIP_CODE_040114.shp')
```

```python
all_shapes = shp.shapes()
```

```python
all_records = shp.records()
```

```python
df11 = pd.read_csv('long_lat_save_11.csv')
```

```python
df11.head()
```

```python
df11.shape
```

```python
df11['zipcode'] = 0
```

```python
df11.astype('int64', inplace=True)
```

```python
for n in tqdm(range(len(df11))):
    for i in range(len(all_shapes)):
        boundary = all_shapes[i]
        if Point(df11['long'][n], df11['lat'][n]).within(shape(boundary)):
            zip_point = all_records[i][0]
            df11['zipcode'][n] = zip_point
            break

df11.to_csv('batch11.csv')
```
